### Select COSMOS images and merge

```python
labels_COSMOS = labels[labels["imaging"]=="COSMOS"]
print("Labels:",len(labels_COSMOS))
metadata_COSMOS = metadata[metadata["imaging"]=="COSMOS"]
print("Metadata:",len(metadata_COSMOS))
```

```python
cols_to_use = list(metadata_COSMOS.columns.difference(labels_COSMOS.columns))
cols_to_use.append("survey_id")
print(cols_to_use)
```

```python
dataset_COSMOS = pd.merge(labels_COSMOS,metadata_COSMOS[cols_to_use],on="survey_id")
dataset_COSMOS
```

### Transform to volunteer votes

```python
hubble_old_pairs = {
    't01_smooth_or_features': ['_a01_smooth_fraction', '_a02_features_or_disk_fraction', '_a03_star_or_artifact_fraction'],
    't02_edgeon': ['_a01_yes_fraction', '_a02_no_fraction'],
    't03_bar': ['_a01_bar_fraction', '_a02_no_bar_fraction'],
    't04_spiral': ['_a01_spiral_fraction', '_a02_no_spiral_fraction'],
    't05_bulge_prominence': ['_a01_no_bulge_fraction', '_a02_just_noticeable_fraction', '_a03_obvious_fraction', '_a04_dominant_fraction'],
    't07_rounded': ['_a01_completely_round_fraction', '_a02_in_between_fraction', '_a03_cigar_shaped_fraction'],
    't09_bulge_shape': ['_a01_rounded_fraction', '_a02_boxy_fraction', '_a03_no_bulge_fraction'],
    't10_arms_winding': ['_a01_tight_fraction','_a02_medium_fraction','_a03_loose_fraction'],
    't11_arms_number': ['_a01_1_fraction', '_a02_2_fraction', '_a03_3_fraction','_a04_4_fraction','_a05_more_than_4_fraction','_a06_cant_tell_fraction'],
    't12_clumpy': ['_a01_yes_fraction', '_a02_no_fraction'],
    't16_clumps_count': ['_a01_1_fraction', '_a02_2_fraction', '_a03_3_fraction', '_a04_4_fraction', '_a05_more_than_4_fraction', '_a06_cant_tell_fraction'],
    't17_clumps_symmetrical': ['_a01_yes_fraction','_a02_no_fraction'],
    't18_clumps_embedded': ['_a01_yes_fraction', '_a02_no_fraction'],
}
```

```python
hubble_new_pairs = {
    'smooth-or-featured-hubble': ['_smooth', '_features', '_artifact'],
    'disk-edge-on-hubble': ['_yes', '_no'],
    'bar-hubble': ['_yes', '_no'],
    'has-spiral-arms-hubble': ['_yes', '_no'],
    'bulge-size-hubble': ['_none', '_just-noticeable', '_obvious', '_dominant'],
    'how-rounded-hubble': ['_completely', '_in-between', '_cigar-shaped'],
    'bulge-shape-hubble': ['_rounded', '_boxy', '_none'],
    'spiral-winding-hubble': ['_tight', '_medium', '_loose'],
    'spiral-arm-count-hubble': ['_1', '_2', '_3', '_4', '_5-plus', '_cant-tell'],
    'clumpy-appearance-hubble': ['_yes', '_no'],
    'clump-count-hubble': ['_1', '_2', '_3', '_4', '_5-plus', '_cant-tell'],
    'galaxy-symmetrical-hubble': ['_yes', '_no'],
    'clumps-embedded-larger-object-hubble': ['_yes', '_no']
}
```

```python
#Which metadata to keep
data_final = dataset_COSMOS[["survey_id","RA","DEC","Z_BEST","MAG_BEST_HI"]]
data_final["filename"] = [str(filename) + ".jpg" for filename in list(data_final["survey_id"])]
data_final
```

```python
for i in range(len(hubble_old_pairs.keys())):
    question_old = list(hubble_old_pairs.keys())[i]
    question_new = list(hubble_new_pairs.keys())[i]
    total_votes = dataset_COSMOS[question_old+"_total_count"]
    for j in range(len(hubble_old_pairs[question_old])):
        answer_fraction = dataset_COSMOS[question_old + hubble_old_pairs[question_old][j]]
        answer_abs = (np.round(answer_fraction * total_votes))
        data_final[question_new + hubble_new_pairs[question_new][j]] = answer_abs
data_final
```

```python
data_final.to_csv("/content/drive/MyDrive/MPE/ZoobotEuclid/Data/gz_hubble_labels_zoobot_complete.csv",index=False)
```
